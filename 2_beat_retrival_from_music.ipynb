```python
import librosa
import IPython.display as ipd
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
```

```python
wave_x, sample_rate = librosa.load('E:/courses/resume_projects/Audio_processing_python/raw_audio_files/58bpm.wav')
ipd.Audio(wave_x, rate=sample_rate)
```

```python
plt.figure(figsize=(15, 4), facecolor=(.9, .9, .9))
librosa.display.waveshow(wave_x, sr=sample_rate, color='red')
```

```python
X = librosa.stft(wave_x)
Xdb = librosa.amplitude_to_db(abs(X))
```

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sample_rate, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
tempo, beat_frames = librosa.beat.beat_track(y=wave_x, sr=sample_rate)

#print('Estimated tempo: {:.2f} beats per minute'.format(tempo))

# Ensure tempo is a scalar, not an ndarray
if isinstance(tempo, (int, float)):  # Check if tempo is a scalar
    print('Estimated tempo: {:.2f} beats per minute'.format(tempo))
else:
    print("Unexpected format for tempo:", tempo)
```

```python
# 4. Convert the frame indices of beat events into timestamps
beat_times = librosa.frames_to_time(beat_frames, sr=sample_rate)
```

```python
print(beat_times)
print(beat_times.shape)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(wave_x, alpha=0.4)
plt.vlines(beat_times, -1, 1, color='r')
plt.ylim(-1, 1)
```
